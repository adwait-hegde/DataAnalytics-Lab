```python
df = df.drop(['Year', 'Month'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Temp' , axis=1)
y = df['Temp']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

Global hypothesis:

H0: 1 = 2 = 3 = 4 = 5 = 0

Ha: 1  2  3  4  5  0


Hypothesis for individual parameters:

H0: i = 0

Ha: i  0


I utilised SAS software to double-check my assumptions because it delivered the statistical data I needed quickly. For my multivariate linear regression model, I utilised the forward elimination strategy to choose variables. The percentage was 5%, or 0.05.

Droped the columns Year and months as it was mentioned in the question

Tried to check the P value with the feature that had highest correlation with the Temp which was CO2 and had a correlation coefficient of 0.75, It showed a p value < 0.0001 so thus we accept it.

Now I am checking with the next highly corelated feature which is N2O but it is observed that the p value obtained using N20 was 0.2274 which is > 0.05 and thus we can remove it from the dataset. It is mainly because there is a high similarity in CO2 and N2O

After rejecting N2O, the highest correlation one can see is 0.0113


On checking the p value of CFC-12, we get to see that it is 0.0158 which is < 0.05 and CH4 0.4182


CO2 0.6059 thus reject and CFC-11 was 0.0002 so accept

TSI 0.0001 so accept


Here, for all the variables, p-values are less than  value of 0.05. Hence, we accept the individual hypothesis.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error ,mean_squared_error, median_absolute_error,confusion_matrix,accuracy_score,r2_score
reg=LinearRegression(fit_intercept=True,normalize=True,copy_X=True, n_jobs=10)
```

```python
cols = ['CFC-12', 'CFC-11', 'TSI', 'MEI', 'Aerosols']
reg.fit(X_train[cols],y_train)
y_pred = reg.predict(X_test[cols])
```

```python
reg.score(X_train[cols], y_train)
```

```python
r2_score(y_test, y_pred)
```

```python
print("Co-efficient: ", reg.coef_)
print("Intercept: ",reg.intercept_)
```

```python
y_pred_LR=reg.predict(X_test[cols])
print(" The Score Linear Regression " ,r2_score(y_test,y_pred_LR)*100)
```

```python
# from scipy import stats
dft = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
variables = ['Aerosols', 'N2O', 'MEI', 'TSI']
X_train, X_test, y_train, y_test = train_test_split(dft[variables], dft['Temp'], test_size=0.2, random_state=42)
LinearRegression_model=LinearRegression(fit_intercept=True,normalize=True,copy_X=True, n_jobs=10)
LR = LinearRegression_model.fit(X_train[variables],y_train)

print("Co-efficient: ", LR.coef_)
print("Intercept: ",LR.intercept_)
```

```python
y_pred_LR=LinearRegression_model.predict(X_test[variables])
print(" The Score Linear Regression " ,r2_score(y_test,y_pred_LR)*100)
```
